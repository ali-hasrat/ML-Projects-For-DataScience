Importing the dependencies

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
````

Data Collection and Preprocessing

````python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
````

````python
print(breast_cancer_dataset)
````

````python
# laoding the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
````

````python
#print first five rows of dataframe
data_frame.head()
````

````python
# no of rows and column in the dataframe
data_frame.shape
````

````python
# loading the 'target column to dataframe
data_frame['label'] = breast_cancer_dataset.target
````

````python
#print last five rows of dataframe
data_frame.tail()
````

````python
# getting some information about data
data_frame.info()
````

````python
# checking the missing value
data_frame.isnull().sum()
````

````python
# statistical measure of data
data_frame.describe()
````

````python
#checking the distribution of Target Value
data_frame['label'].value_counts()
````

    1 --> Bening
    0 --> Malignant

````python
data_frame.groupby('label').mean()
````

Separating the features and Target

````python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
````

````python
print(X)
````

````python
print(Y)
````

Splitting the data into training and testing data

````python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

Standarized the Data

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
````

````python
print(X_train_std )
````

````python
print(X_test_std)
````

**Building the Neural Network**

Neural-Networks-Architecture.png

````python
# importing the tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
````

````python
# setting up the layers of Neural Network

model = keras.Sequential([

                keras.layers.Flatten(input_shape=(30,)),
                keras.layers.Dense(20,activation='relu'),
                keras.layers.Dense(2,activation='sigmoid')
])
````

````python
# compiling the Neural Network

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
)
````

````python
# training the Neural Network

history = model.fit(X_train, Y_train, validation_split=0.1, epochs=12)
````

Visualizing accuracy and Loss


````python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_acccuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc='lower right')
````

````python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc='upper right')
````

Accuracy of the model on the test data

````python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
````
